### Import Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.figure import Figure
import matplotlib.ticker as mtick
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
```

### Input data

```python
data = pd.read_excel('change in house values.xlsx',0)
```

### Trip data

```python
#by default, the dataframe's index is set to be series of number, but in this case setting it to cities makes more sense.
data.set_index('Cities', inplace=True)
```

### Creating variables for triple bar graphs

```python
#by default, multi-bar graphs are overlapping each other. 
#We have to create a series of numbers for x indexes, so we can offset bars one another.
x_indexes = np.arange(len(data["Past Month"]))
x_indexes2 = np.arange(len(data["Past 3 Month"]))
x_indexes3 = np.arange(len(data["Past 12 Month"]))
width = 0.37
```

### Plot graphs

```python
#ploting 3 graphs -Past Month, Past 3 Months, and Past 12 Months

plt.bar(x_indexes,data["Past Month"], width = 0.25,color="#54457F",align='edge')
plt.bar(x_indexes2+ width,data["Past 3 Month"], width = 0.25, color = "#222E50")
plt.bar(x_indexes3+ width+0.22,data["Past 12 Month"], width = 0.25,color = "#007991")

#referencing the source
plt.text(3.3, 8.2, "Source: Hobart Property Report Q3 2020, Nest Property, 2020", fontsize=10)

#structring the graph
plt.title('Change in Properties Value\n',fontsize=21, fontweight ="bold",fontname= 'DejaVu Sans')
plt.xticks(ticks = x_indexes+0.07, labels= data.index,fontsize= 12, ha="left")
plt.yticks(fontsize= 13)
plt.xlabel('\nCities', fontdict = {'fontname': 'DejaVu Sans', 'fontsize': 16})
plt.ylabel('Change in Value\n', fontdict = {'fontname': 'DejaVu Sans', 'fontsize': 16})
plt.xlim(0,8)
plt.legend(["Past Month","Past 3 Month","Past 12 Month"], loc =4,labelcolor='linecolor')

#styling the graph
plt.tight_layout()

#The gcf() function in pyplot module of matplotlib library is used to get the current Axes instance on the current figure matching the given keyword args, or create one.
fig = plt.gcf()
fig.set_facecolor('#E5ECF4')
fig.figsize=(10,15)

#get current axes
ax = plt.gca()
ax.set_facecolor('#E5ECF4')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(decimals=0))

plt.show()

```
